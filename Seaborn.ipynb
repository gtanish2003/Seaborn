```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

# bar plot
### when there are multiple observations in each category, it also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using error bars 

```python
sns.barplot(data=tips, x='sex',y='total_bill')
```

```python
sns.catplot(data=tips,x='sex',y='total_bill',kind='bar',errorbar=None)
```

# it is giving average by default ...but we can calculate other functions as well
### for this we have a parameter known as estimator

```python
sns.barplot(data=tips,x='sex',y='total_bill',estimator=np.median)
```

```python
# pointplot  
### it is similar to barplot minus the bars
### it is used to represent the point estimate and confidence interval

sns.pointplot(data=tips,x='sex',y='total_bill',estimator=np.mean)
```

```python
# countplot
### it is similar to barplot except that the estimator is explicitly counting the number of occurences
sns.countplot(data=tips,x='sex')
```

```python
sns.countplot(data=tips,x='day',hue='sex')
```

```python
# we can also perform faceting like other figure level functions
sns.catplot(data=tips,x='sex',hue='day',kind='count',col='smoker')
```

# 5. Regression Plots

```python
iris
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```

```python
# PairGrid

g=sns.PairGrid(data=iris)
g.map(sns.scatterplot)
```

```python
g=sns.PairGrid(data=iris)
g.map(sns.histplot)
```

```python
g=sns.PairGrid(data=iris)
g.map_diag(sns.boxplot)
g.map_offdiag(sns.scatterplot)
```

### Pairplot is a special case of PairGrid where the diagonal plots are either histogram or kde plot and non diagonal plots are scatter plot.

```python
g=sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.boxplot)
g.map_lower(sns.scatterplot)
```

```python
# we can also restrict the columns onto which we want to work
# for this there exist a parameter called vars
g=sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.boxplot)
g.map_lower(sns.scatterplot)
```

# JointPlot vs JointGrid

```python
# there exist a parameter called kind -- its default value is scatter plot but it can be given any type according to the needs

sns.jointplot(data=tips,x='total_bill',y='tip',kind='kde')
```
